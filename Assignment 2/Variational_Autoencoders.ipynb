## Creating Dataloader for Autoencoders

```python
class CityDataset(Dataset):
    def __init__(self, dataset, transform):
        self.dataset = dataset
        self.transform = transform
        self.len_dataset = len(dataset)
        self.label_model = label_model

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self , idx):
        img = Image.open(self.dataset[idx]).convert('RGB')
        cityscape, label = split_image(img)
        cityscape = self.transform(cityscape)
        label_class = self.label_model.predict(label.reshape(-1,3)).reshape(256,256)
        label_class = torch.Tensor(label_class).long()
        return cityscape, label_class
        
transforms = T.Compose([
    T.ToTensor(),
    T.Normalize(mean = (0.485, 0.456, 0.406), std = (0.229,0.224,0.225))
])

dataset_t = CityDataset(images , transforms)     
dataset_v = CityDataset(vimages , transforms)

BATCH_SIZE = 64
NUM_WORKER = 2
train_loader = DataLoader(
    dataset=dataset_t,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKER,
    shuffle=True
)
valid_loader = DataLoader(
    dataset=dataset_v,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKER,
    shuffle=True
)
```

# Defining EncoderDecoder Model
* Here I used Resnet18 model for Encoder and similar Reversed Model as Decoder which Return Final Image with Channel lenght as number of Clusters.

```python
class EncoderDecoder(nn.Module):
    def __init__(self, encoded_image_size = 1):
        super(Encoder, self).__init__()
        self.out_dim = 512
        self.n_class = 21
        self.enc_image_size = encoded_image_size
        net = torchvision.models.resnet18(pretrained=True)
        modules = list(net.children())[:-2]
        self.resnet = nn.Sequential(*modules)
        self.flatten = nn.Flatten()
        self.adaptive_pool = nn.AdaptiveAvgPool2d((encoded_image_size, encoded_image_size))
        self.fine_tune()
        
        self.relu = nn.ReLU(inplace = True)
        self.deconv1 = nn.ConvTranspose2d(512, 512, kernel_size=3, stride = 2, padding= 1, dilation = 1, output_padding= 1)
        self.bn1 = nn.BatchNorm2d(512)
        self.deconv2 = nn.ConvTranspose2d(512, 256, kernel_size=3, stride = 2, padding= 1, dilation = 1, output_padding= 1)
        self.bn2 = nn.BatchNorm2d(256)
        self.deconv3 = nn.ConvTranspose2d(256, 128, kernel_size=3, stride = 2, padding= 1, dilation = 1, output_padding= 1)
        self.bn3 = nn.BatchNorm2d(128)
        self.deconv4 = nn.ConvTranspose2d(128, 64, kernel_size=3, stride = 2, padding= 1, dilation = 1, output_padding= 1)
        self.bn4 = nn.BatchNorm2d(64)
        self.deconv5 = nn.ConvTranspose2d(64, 32, kernel_size=3, stride = 2, padding= 1, dilation = 1, output_padding= 1)
        self.bn5 = nn.BatchNorm2d(32)
        self.classifier = nn.Conv2d(32, self.n_class, kernel_size= 1)
        
    def get_out_dim(self):
        return self.out_dim

    def forward_once(self, images):
        out = self.resnet(images)
        out = self.adaptive_pool(out)
        out = out.permute(0, 2, 3, 1)
        out = out.view(out.size(0), -1, out.size(-1))
        out = self.flatten(out)
        return out
    
    def forward(self , image):
        out = self.resnet(image)
        x = self.bn1(self.relu(self.deconv1(out)))
        x = self.bn2(self.relu(self.deconv2(x)))
        x = self.bn3(self.relu(self.deconv3(x)))
        x = self.bn4(self.relu(self.deconv4(x)))
        out_decoder = self.bn5(self.relu(self.deconv5(x)))
        score = self.classifier(out_decoder)
        return score
    
    def fine_tune(self, last_x = 1,fine_tune=True):
        for p in self.resnet.parameters():
            p.requires_grad = False
        for c in list(self.resnet.children())[-last_x]:
            for p in c.parameters():
                p.requires_grad = fine_tune
```
